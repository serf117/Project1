```python
import pandas as pd
```

```python
#DISCLAIMER: the data i used fo this excercise is from the sample file, result might be different if the source is changed to the google drive file(20gb)
#importing data
data = pd.read_csv("sample.csv")
```

```python
data
```

```python
#number of unique monitored commercial chains
n=len(pd.unique(data['cadenaComercial']))
print(n)
#names of the monitores commercial chains
chainarr=pd.unique(data['cadenaComercial'])
for chain in chainarr:
    print(chain)
```

```python
#top 10 monitored products
s =data['producto'].value_counts()
s[:10]
```

```python
# commercial chain with the highest number of monitored products
data.groupby(['cadenaComercial']).count().nlargest(4,'producto')
```

```python
# commercial chain with the highest number of unique monitored products
data.groupby(['cadenaComercial']).agg({"producto": "nunique"}).reset_index().nlargest(4,'producto')
```

```python
# township/municipality with most monitored products
data.groupby(['municipio']).count().nlargest(10,'producto')
```

```python
#top 10 monitored brands
data.groupby(['marca']).count().nlargest(10,'producto')
```
